Question_1:Find the cube of a given number using def and lambda function both 

```python
def cube(x):
    return x*x*x
cubiodic=lambda x:x*x*x
print(cube(3))
print(cubiodic(4))
```

List comprehension :

Question_2:cube of the range from 1 to 10 using the list comprehension?


```python
cube=[x*x*x for x in range(1,11)]
print(cube)
```

Question_3:Cube of the range from 1 to 10 only the even number using the list comprehension?

```python
cube=[x*x*x for x in range(1,11) if x%2==0]
print(cube)
```

Python filter function

Filter function : is generally used for the filtering the some element which is required by the condition.

```python
def even(n):
    return n%2==0
a=[1,2,3,4,5,6]
b=filter(even,a)
print(list(b))
```

Python Map function   
Map function: this is function do is perform function to every element of the iteratable.

```python
def square(n):
    return n*n
a=[12,3,2,23,4,24,22,32]
b=map(square,a)
print(list(b))
```

Python Reduce function   
Reduce Function:it will take the element from the list and then shrink that to one value.


```python
from functools import reduce
def add(x,y):
    return x+y
a=[112,12,12,12,1]
re=reduce(add,a)
print(re)
```
