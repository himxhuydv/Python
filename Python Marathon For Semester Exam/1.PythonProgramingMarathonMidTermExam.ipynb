```python
#1️ Grade system

# Input: Marks

# 90+ → A

# 75–89 → B

# 50–74 → C

# Below 50 → Fail
Marks=int(input("enter your marks "))
if(Marks>=0):
    if(Marks>=90):
        print("A")
    elif(Marks>=75 and Marks<=89):
        print("B")
    elif(Marks>=50 and Marks<=74):
        print("C")
    else:
        print("Fail")
else:
    print("Negative Input")
```
