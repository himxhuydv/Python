# Constructor



it's automatically initial the special method.


```python
class faculty:
    def __init__(self):
        self.id=int(input("Enter the number"))
        self.name=input("Enter the name")
        self.salary=float(input("Enter the salary"))

    def display(self):
        print(self.id)
        print(self.name)
        print(self.salary)
a=faculty()
a.display()
print(a.salary)
```
