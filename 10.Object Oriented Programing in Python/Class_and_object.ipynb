## Simple class and object example 

```python
class employee: # here we are creating the class first 
    def putdetail(self): # Method or function what will our class will do (actually take input )
        self.id=int(input("Enter the employee id "))
        self.name=input("Enter the Name")
        self.salary=float(input("Enter the salary"))
    def display(self): # Print everything out what needs to print 
        print("Your employee id",self.id)
        print("Your employee name",self.name)
        print("Your employee Salary",self.salary)
#creating the object now 
a=employee()
a.putdetail()
a.display()
```
